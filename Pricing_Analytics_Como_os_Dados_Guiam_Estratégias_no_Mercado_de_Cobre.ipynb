<h1>Pricing Analytics: Como os Dados Guiam Estratégias no Mercado de Cobre


<h2>Introdução

A precificação do cobre é um fator estratégico essencial para a competitividade e sustentabilidade das empresas do setor de materiais elétricos, sendo diretamente influenciada pela volatilidade do mercado de commodities. O cobre é um insumo fundamental para a fabricação de cabos, fios e componentes elétricos, e suas oscilações de preço impactam diretamente os custos de produção e a precificação dos produtos finais. Empresas que dependem desse metal precisam adotar abordagens analíticas para prever tendências de mercado e estruturar estratégias de precificação mais eficientes, minimizando riscos financeiros e garantindo margens de lucro sustentáveis.


De acordo com a London Metal Exchange (LME), os preços oficiais são a referência global para contratos físicos de metais, sendo amplamente utilizados por produtores, consumidores e comerciantes para precificação e liquidação de transações comerciais. Esses preços são determinados por meio de negociações em sessões oficiais do mercado, refletindo as dinâmicas da oferta e demanda internacional (LME, 2024). Tais oscilações reforçam a necessidade do uso de abordagens baseadas em Business Intelligence (BI) e Analytics, que permitem uma tomada de decisão mais dinâmica e orientada por dados.

Kotler e Keller (2006, p. 429) destacam que a determinação de preços é uma das decisões mais complexas para as empresas, pois envolve não apenas os custos internos, mas também a concorrência, as condições do mercado e a percepção de valor pelos clientes. Segundo os autores, “muitas empresas acabam adotando ‘estratégias’ como ‘determinar os custos internos e aplicar margens tradicionais do setor’, sem levar em consideração os fatores externos que influenciam a precificação” .

A aplicação de modelos preditivos de machine learning, como Regressão Linear e Random Forest, possibilita a identificação de padrões históricos e a antecipação de movimentos futuros nos preços do cobre, oferecendo suporte à tomada de decisão estratégica. Além disso, a Análise de Sensibilidade possibilita avaliar o impacto dessas oscilações nos custos de produção e na rentabilidade dos produtos, permitindo ajustes mais precisos nas estratégias de precificação.

Diante desse cenário, este estudo propõe uma abordagem baseada em dados históricos da LME, aplicando técnicas de machine learning e análise de sensibilidade para auxiliar empresas do setor de materiais elétricos na formulação de estratégias de precificação mais eficazes. A pesquisa busca contribuir para a literatura acadêmica e empresarial ao demonstrar como a inteligência analítica pode transformar a tomada de decisão em mercados altamente voláteis, promovendo maior estabilidade financeira e competitividade para as organizações.

Vamos utilizar Pandas para fazer uma análise preliminar da nossa base de dados, verificar as primeiras linhas da nossa base de dados, o número de linhas e colunas, e os tipos de dados de cada coluna.

Isso nos ajudará a confirmar se o carregamento foi bem-sucedido e se as colunas 'Date', 'Value' e 'Currency' foram identificadas corretamente.

```python
import pandas as pd
from IPython.display import display

try:
    df_copper = pd.read_csv('copper-prices-historical-chart-data.csv', parse_dates=['date'])
    display(df_copper.head())
    print(df_copper.shape)
    print(df_copper.dtypes)
except FileNotFoundError:
    print("Error: 'copper-prices-historical-chart-data.csv' not found.")
    df_copper = None
except pd.errors.ParserError:
    print("Error: Could not parse the CSV file correctly. Check delimiter and file format.")
    df_copper = None
except Exception as e:
    print(f"An unexpected error occurred: {e}")
    df_copper = None
```

Aqui podemos ver que nossa base de dados contém 16520 linhas e três colunas, sendo date (data), value (valor por libra em dólares) e currency (moeda, dólares americanos). Também podemos verificar os tipos de dados das nossas variáveis, sendo respectivamente 'datetime64', 'float64'e 'object'. Para o tratamento de dados, nós criaremos um subdataset, com os valores dos últimos 5 anos. Porém, para efeito de comparação, manteremos o nosso dataset principal, pois assim teremos como comparar os resultados da aprendizagem de máquina, com dados históricos de 45 anos, com o resultado de dados históricos dos últimos 5 anos.

```python
if df_copper is not None:
    # Define the start date for the last 5 years (since 2020)
    start_date = '2020-01-01'
    # Filter the DataFrame to include only data from the last 5 years
    df_copper_5years = df_copper[df_copper['date'] >= start_date].copy()
    print("\nDataFrame for the last 5 years:")
    display(df_copper_5years.head())
    print(df_copper_5years.shape)
    print(df_copper_5years.dtypes)
else:
    print("df_copper was not loaded correctly, cannot create df_copper_5years.")
```

Nosso subdataset contém 1381 linhas, as mesmas 3 colunas, e contém dados a partir de Janeiro de 2020.

```python
# Check the data type of the 'Value' column
print(df_copper['value'].dtypes)

# Attempt to convert 'Value' column to numeric, handling potential errors
try:
    df_copper['value'] = pd.to_numeric(df_copper['value'])
    print("\n'Value' column successfully converted to numeric.")
except ValueError as e:
    print(f"\nError converting 'Value' column to numeric: {e}")
    print("Please check the 'Value' column for non-numeric characters (e.g., commas, currency symbols) and clean the data.")

# Descriptive statistics for the 'Value' column
if pd.api.types.is_numeric_dtype(df_copper['value']):
    print("\nDescriptive Statistics for Copper Prices (Value):")
    print(df_copper['value'].describe())
else:
    print("\nCould not perform descriptive statistics as 'Value' column is not numeric.")
```

Uma forma interessante de representar visualmente a variação dos dados para os dois datasets (últimos 45 anos e últimos 5 anos) e destacar as diferenças nas estatísticas descritivas, como a média e a distribuição, seria utilizando um box plot ou violin plot.

Ambos os tipos de gráficos são eficazes para visualizar a distribuição de dados e comparar distribuições entre diferentes grupos (neste caso, os dois períodos de tempo).

Box Plot: Um box plot mostraria a mediana, quartis (25º e 75º percentis), a faixa interquartil (IQR) e possíveis outliers para cada dataset. Isso permitiria ver claramente:

Onde está a mediana de cada período.
A dispersão dos dados (pela largura da caixa).
A faixa geral dos preços (pelos "bigodes").
A presença de outliers.
Comparando os box plots dos dois períodos, podemos analisar visualmente a diferença na mediana (bem maior nos últimos 5 anos) e na dispersão (menor nos últimos 5 anos).

Violin Plot: Um violin plot é semelhante a um box plot, mas também mostra a densidade de probabilidade dos dados em diferentes valores. Ele fornece uma visão mais rica da distribuição, mostrando se há múltiplos picos ou outras características na distribuição dos preços.

Comparando os violin plots, podemos ver a mediana e a dispersão, mas também a forma da distribuição dos preços em cada período. Por exemplo, se a distribuição nos últimos 5 anos for mais concentrada em torno de valores mais altos, o violin plot mostra isso claramente.

```python
import matplotlib.pyplot as plt
import pandas as pd

# Criar uma lista com os dados de preço dos dois datasets
data_to_plot = [df_copper['value'].dropna(), df_copper_5years['value'].dropna()]

# Criar o box plot
plt.figure(figsize=(8, 6))
plt.boxplot(data_to_plot, labels=['Últimos 45 Anos', 'Últimos 5 Anos'], patch_artist=True)

# Adicionar título e rótulos
plt.title('Distribuição dos Preços do Cobre: Comparativo Histórico vs. Últimos 5 Anos')
plt.ylabel('Preço do Cobre (USD por Libra)')
plt.grid(axis='y', linestyle='--')

# Mostrar o gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Assumindo que df_copper e df_copper_5years já estão carregados e limpos

# Criar uma lista com os dados de preço dos dois datasets
data_to_plot = [df_copper['value'].dropna(), df_copper_5years['value'].dropna()]

# Criar o violin plot
plt.figure(figsize=(8, 6))
plt.violinplot(data_to_plot, showmeans=False, showmedians=True) # showmedians=True para exibir a mediana

# Configurar os rótulos do eixo x
plt.xticks([1, 2], ['Últimos 45 Anos', 'Últimos 5 Anos'])

# Adicionar título e rótulos
plt.title('Distribuição dos Preços do Cobre: Comparativo Histórico vs. Últimos 5 Anos')
plt.ylabel('Preço do Cobre (USD por Libra)')
plt.grid(axis='y', linestyle='--')

# Mostrar o gráfico
plt.show()
```
